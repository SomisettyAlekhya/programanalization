```python
def find_consecutive_work(df):
    consecutive_days = 0
    prev_employee = None

    for index, row in df.iterrows():
        current_employee = row['Employee Name']
        
        if current_employee == prev_employee:
            consecutive_days += 1
        else:
            consecutive_days = 1
        
        if consecutive_days == 7:
            print(f"Employee Name: {current_employee}, Position ID: {row['Position ID']} - Worked for 7 consecutive days")
        
        prev_employee = current_employee

# Task 2b: Print the name and position of employees who have less than 10 hours between shifts but greater than 1 hour.
def find_short_breaks(df):
    for index, row in df.iterrows():
        current_time = row['Time']
        prev_row = df.iloc[index - 1] if index > 0 else None

        if prev_row is not None:
            time_between_shifts = current_time - prev_row['Time Out']
            
            if 1 < time_between_shifts.total_seconds() / 3600 < 10:
                print(f"Employee Name: {row['Employee Name']}, Position ID: {row['Position ID']} - Has short breaks between shifts")

# Task 2c: Print the name and position of employees who have worked for more than 14 hours in a single shift.
def find_long_shifts(df):
    for index, row in df.iterrows():
        shift_duration = row['Time Out'] - row['Time']
        
        if shift_duration.total_seconds() / 3600 > 14:
            print(f"Employee Name: {row['Employee Name']}, Position ID: {row['Position ID']} - Worked more than 14 hours in a single shift")

# Call the functions to analyze the data
find_consecutive_work(df)
find_short_breaks(df)
find_long_shifts(df)
```

```python
consecutive_work_results = []
short_breaks_results = []
long_shifts_results = []

# Task 2a: Collect the name and position of employees who have worked for 7 consecutive days.
def find_consecutive_work(df):
    consecutive_days = 0
    prev_employee = None

    for index, row in df.iterrows():
        current_employee = row['Employee Name']
        
        if current_employee == prev_employee:
            consecutive_days += 1
        else:
            consecutive_days = 1
        
        if consecutive_days == 7:
            consecutive_work_results.append({'Employee Name': current_employee, 'Position ID': row['Position ID']})
        
        prev_employee = current_employee

# Task 2b: Collect the name and position of employees who have less than 10 hours between shifts but greater than 1 hour.
def find_short_breaks(df):
    for index, row in df.iterrows():
        current_time = row['Time']
        prev_row = df.iloc[index - 1] if index > 0 else None

        if prev_row is not None:
            time_between_shifts = current_time - prev_row['Time Out']
            
            if 1 < time_between_shifts.total_seconds() / 3600 < 10:
                short_breaks_results.append({'Employee Name': row['Employee Name'], 'Position ID': row['Position ID']})

# Task 2c: Collect the name and position of employees who have worked for more than 14 hours in a single shift.
def find_long_shifts(df):
    for index, row in df.iterrows():
        shift_duration = row['Time Out'] - row['Time']
        
        if shift_duration.total_seconds() / 3600 > 14:
            long_shifts_results.append({'Employee Name': row['Employee Name'], 'Position ID': row['Position ID']})

# Call the functions to collect the results
find_consecutive_work(df)
find_short_breaks(df)
find_long_shifts(df)

# Create DataFrames from the collected results
consecutive_work_df = pd.DataFrame(consecutive_work_results)
short_breaks_df = pd.DataFrame(short_breaks_results)
long_shifts_df = pd.DataFrame(long_shifts_results)

# Display the results DataFrames
print("Employees with 7 consecutive days of work:")
print(consecutive_work_df)

print("\nEmployees with short breaks between shifts:")
print(short_breaks_df)

print("\nEmployees with shifts longer than 14 hours:")
print(long_shifts_df)
```
